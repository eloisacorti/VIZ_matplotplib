###**EXERCISE 12**: ***



Plot required data as shown below:
>$y_1=x^3*e^{-x^2}$

>$y_2=x^4*e^{-x^2}$

>$y_3=x^5*e^{-x^2}$


```
import math
x = np.arange (0,3, 0.05)
y1 = x **3 * np.exp(-x**2)
y2 = x **4 * np.exp(-x**2)
y3 = x **5 * np.exp(-x**2)
```

```
plt.figure(figsize=[12,6])
plt.plot(x,y1, ls='-.', color='b', label=r'$y_1= x^{3}*e^{-x^{2}}$')
plt.plot(x,y2, color='C9',drawstyle='steps',label=r'$y_2= x^{4}*e^{-x^{2}}$' )
plt.plot(x,y3, marker='^', ls='None',label=r'$y_3= x^{5}*e^{-x^{2}}$')
plt.title('MATPLOTLIB PLOTTING')
plt.grid(color= 'C9')
plt.legend( bbox_to_anchor=[1,0.15]);
```

###**EXERCISE 14**: ***



Plot the following grid of plots where:

```
x = np.linspace(0, 2, 25)
fig, axs = plt.subplots(3,1,figsize=[8,12])
plt.xlabel('x')


i=0
colores = ['C9','C4','r']
estilo_linea = ['--','-','-.']
for ax in axs:
  ax.plot(x,x**i, label=f'$x^{i}$', color = colores[i], ls = estilo_linea[i])
  ax.set_title("Plot nº "+ str(i))
  ax.legend(loc=2)
  plt.setp(axs[i], ylabel='y')
  i = i + 1 
```

Plot the following axeses with provided data:

###**EXERCISE 18**: ****



Using the data set available at [https://raw.githubusercontent.com/immune-institute/viz-datasets/main/wines_reviews.csv](https://raw.githubusercontent.com/immune-institute/viz-datasets/main/wines_reviews.csv) create the bar plot shown below:

```
!gdown  https://raw.githubusercontent.com/immune-institute/viz-datasets/main/wines_reviews.csv
```

```
import pandas as pd
df_wine = pd.read_csv('/content/wines_reviews.csv').set_index('Unnamed: 0')
```

```
point_df= df_wine.filter(['points', 'description']).groupby('points').count().reset_index()
variety_df=df_wine.filter(['variety', 'description']).groupby('variety').count().reset_index().sort_values('description', ascending = False).head(20)
origin_df = df_wine.filter(['province', 'description']).groupby('province').count().reset_index().sort_values('description', ascending = False).head(20)
```

```
fig, axs = plt.subplots(2,2,figsize = [16,8])
axs[0,0].bar('points', 'description', data = point_df, color = 'mediumvioletred',width = 0.5)
axs[0,0].set_title('Wine Scores', size = 20)
axs[0,1].hist('price', data =df_wine,color = 'mediumvioletred' )
axs[0,1].set_title('Wine Prices', size = 20)
axs[1,0].bar('variety', 'description', data = variety_df, color = 'mediumvioletred')
axs[1,0].set_xticklabels(variety_df.variety,rotation = 90)
axs[1,0].set_title('Wine Varieties', size = 20)
axs[1,1].bar('province', 'description', data = origin_df, color = 'mediumvioletred')
axs[1,1].set_xticklabels(origin_df.province,rotation = 90)
axs[1,1].set_title('Wine Origins', size = 20);
```
